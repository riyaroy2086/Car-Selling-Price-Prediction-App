```python
final_dataset=df[['Year','Selling_Price','Present_Price','Kms_Driven','Fuel_Type','Seller_Type','Transmission','Owner']]
```

```python
final_dataset.head()
```

```python
final_dataset['Current Year'] = 2022
```

```python

# Converting the categorical features into one-hot encoded

final_dataset = pd.get_dummies(final_dataset,drop_first= True)

#drop_first= True -> 1st col to be deleted so to prevent the tble from dummy variable trap
```

```python
X=final_dataset.iloc[:,1:] # excluding the selling price index, all others are independent variables
Y=final_dataset.iloc[:,0] # selling price is the dependent variable
```

```python
X.head()
```

```python
Y.head()
```

```python
# Feature Importance
from sklearn.ensemble import ExtraTreesRegressor
# so as to understand which are the main features from so many given features
model=ExtraTreesRegressor()
model.fit(X,Y)
```

```python
print(model.feature_importances_)
```

```python
#Plotting the importance categories graph
feat_importances=pd.Series(model.feature_importances_,index=X.columns)
#creating a series using pd for bthe values listed above :
#[0.18907507 0.00471789 0.12071376 0.30697528 0.00074573 0.15575027 0.222022  ]
feat_importances.nlargest(5).plot(kind='pie')
plt.show()
```

```python
#Train-Test data Split
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
```

```python
X_train.shape
```

```python
from sklearn.ensemble import RandomForestRegressor
rf_random=RandomForestRegressor()
```

```python
#Hyper parameters
n_estimators=[int(x) for x in np.linspace(start=100,stop=1200,num=12)]
print(n_estimators)   # diff decision trees for my hyper parameters 
```

```python
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
```

```python
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(5, 30, num = 6)]
```

```python
# max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10, 15, 100]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 5, 10]
```

```python
# Creating the random grid
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf}

print(random_grid)
```

```python
# Use the random grid to search for best hyperparameters
# First create the base model to tune
rf = RandomForestRegressor()
```

```python
# Random search of parameters, using 3 fold cross validation, 
# search across 100 different combinations
rf_random = RandomizedSearchCV(estimator = rf, param_distributions = random_grid,scoring='neg_mean_squared_error', n_iter = 10, cv = 5, verbose=2, random_state=42, n_jobs = 1)
```

```python
rf_random.fit(X_train,Y_train)
```

```python
#Creating the prediction for X_test data
predictions=rf_random.predict(X_test)
```

```python
sns.displot(Y_test-predictions)
```

```python
plt.scatter(Y_test,predictions)
```

```python
from sklearn import metrics
print('MAE:', metrics.mean_absolute_error(Y_test, predictions))
print('MSE:', metrics.mean_squared_error(Y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(Y_test, predictions)))
```
